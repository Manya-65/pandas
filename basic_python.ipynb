```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.hist(df1['SAT'])
plt.xlabel('SAT scores')
plt.ylabel('Frequency')
plt.title('distrubtion')
plt.show()
```

```python
sns.histplot(df1['SAT'],color='pink')
plt.show
```

```python
plt.plot(df1['SAT'],df1['Accept'],'ro')
plt.xlabel("SAT scores")
plt.ylabel("Frequency")
plt.title("distrubtion")
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Top10'])
plt.xlabel('SAT scores')
plt.ylabel('Top10 percentage')
plt.title('SAT scores vs. Top10 percentage')
plt.show()
```

```python
sns.pairplot(df1)
plt.show()
```

```python
sns.pairplot(df1.iloc[:,1:3])
plt.show()
```

```python
plt.boxplot(df1['SAT'])
plt.show()
```
